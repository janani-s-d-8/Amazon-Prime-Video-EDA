Bivariate Visualizations

```python
# Bivariate 1: Type vs Release Year (Post-2010)
plt.figure(figsize=(12, 6))
recent_titles = titles_df[titles_df['release_year'] >= 2010]
sns.countplot(data=recent_titles, x='release_year', hue='type', palette='muted')
plt.title("Number of Titles by Type per Year (2010+)")
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title='Type')
plt.show()

# Bivariate 2: Type vs Seasons (TV Shows only)
tv_shows = titles_df[titles_df['type'] == 'SHOW']
plt.figure(figsize=(10, 6))
sns.boxplot(data=tv_shows, x='type', y='seasons', palette='coolwarm')
plt.title("Distribution of Seasons in TV Shows")
plt.ylabel("Number of Seasons")
plt.xlabel("Type")
plt.show()

# Bivariate 3: Type vs IMDb Score
plt.figure(figsize=(10, 6))
sns.violinplot(data=titles_df, x='type', y='imdb_score', palette='Set1')
plt.title("IMDb Score Distribution by Type")
plt.ylabel("IMDb Score")
plt.xlabel("Type")
plt.show()
```

Insights from Content Creation & Viewership Trends

Q1: Which years saw the most content release?


```python
content_by_year = titles_df['release_year'].value_counts().sort_index()
plt.figure(figsize=(14, 6))
sns.lineplot(x=content_by_year.index, y=content_by_year.values, marker='o', color='purple')
plt.title("Total Content Released per Year")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.grid(True)
plt.show()
```

Q2: What genres dominate content creation?

```python
top_genres = titles_df['genres'].str.split(', ').explode().value_counts().head(10).reset_index()
top_genres.columns = ['genre', 'count']

plt.figure(figsize=(10, 6))
sns.barplot(data=top_genres, y='genre', x='count', hue='genre', palette='rocket', dodge=False, legend=False)
plt.title("Top 10 Most Frequent Genres")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.grid(axis='x')
plt.show()
```

Q3: Which countries contribute the most content?

```python
top_countries = titles_df['production_countries'].str.split(', ').explode().value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(y=top_countries.index, x=top_countries.values, palette='mako')
plt.title("Top 10 Content-Producing Countries")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```

IMDb Score Distribution over Time

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=titles_df, x='release_year', y='imdb_score', hue='type', alpha=0.6, palette='Set2')
plt.title("IMDb Score vs Release Year")
plt.xlabel("Release Year")
plt.ylabel("IMDb Score")
plt.grid(True)
plt.legend(title='Type')
plt.show()
```
